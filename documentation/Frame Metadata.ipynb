This tells us a few things:
- The skeleton and contour is NaN iffdef `is_nan_skeleton == False` iff `Segementation Status == 's'` iff `Frame Code == 1`
- `Segmentation Status == 'm'` iff `Frame Code == 2`
- `Segmentation Status == 'd'` iff `Frame Code == 3`
- `Segmentation Status == 'f'` iff `Frame Code >= 100` (i.e. if Frame Code is in one of the failure modes)
- `Segementation Status == 'n'` never appears

From this I conclude that we can drop Segmentation Status in favour of an `@property`-decorated method of NormalizedWorm that uses `frame_code`:

```Python
class NormalizedWorm:

    #...
    
    @property
    def segmentation_status(self):
        try:
            return self._segmentation_status
        except AttributeError:
            s = self.frame_code == 1
            m = self.frame_code == 2
            d = self.frame_code == 3

            self._segmentation_status = np.empty(self.num_frames, dtype='<U1')
            for frame_index in range(self.num_frames):
                if s[frame_index]:
                    self._segmentation_status[frame_index] = 's'
                elif m[frame_index]:
                    self._segmentation_status[frame_index] = 'm'
                elif d[frame_index]:
                    self._segmentation_status[frame_index] = 'd'
                else:
                    self._segmentation_status[frame_index] = 'f'
        
            return self._segmentation_status        
```

This result is confusing since we have a line of code in [worm_features.py](https://github.com/openworm/open-worm-analysis-toolbox/blob/master/open-worm-analysis-toolbox/features/worm_features.py) that defines:
```
is_stage_movement = nw.segmentation_status == 'm'
```
But here, with the loaded example data, it appears the Schafer lab code created a contour `.mat` file that has `is_stage_movement == 1` for 105 instances where `Segmentation Status` == 'd', for dropped frame!  I'm not sure how this is possible.  The field `is_stage_movement = segmentation_status == 'm'` is passed for use in calculating `locomotion.turns`.  This is its only use for feature calculation.

Looking at the [SegWorm codebase](https://github.com/openworm/SegWorm), I can find no reference to `is_stage_movement` or `segmentation_status`, so looking back at the old code is not helpful.  I can't understand how those old .mat files were even generated with that field name if it does not appear in the body anywhere.

If all the frames the contour file is claiming are stage movements are truly stage movements, then we are currently overstating the number of omega and upsilon turns.  Therefore there will be a discrepancy between a NormalizedWorm from Schafer file and a NormalizedWorm derived from a BasicWorm loaded from a Schafer file:

````python
# Prepare the labels
frame_code_names = dict(zip(frame_code_info['Frame Code'], frame_code_info['Code Name']))
codes_actually_used = np.unique(nw.frame_code)
labels_needed = list(frame_code_names[d] for d in codes_actually_used)

# Create a pie chart
frame_counts = {i: np.bincount(nw.frame_code)[i] for i in np.unique(nw.frame_code)}
patches, texts, autotexts = plt.pie(x=list(frame_counts.values()),
                                    explode=tuple([0,0,0,0] + [1.55]*(len(frame_counts)-4)),
                                    labels=labels_needed,
                                    autopct='%1.1f%%',
                                    startangle=15,
                                    colors=['g', 'r', 'c', 'y', 'm'], 
                                    labeldistance=1.2)
plt.title("Proportion of frames segmented")
plt.show()
````
